### Creates a wait for download function

```python
def wait_for_downloads():
    print("Waiting for downloads", end="")
    while any([filename.endswith(".crdownload") for filename in
               os.listdir("/Users/emily/Downloads/")]):
        time.sleep(2)
        print(".", end="")
    print("done!")
```

## Download the images

```python
import os

urlfiles = {'url': 'file'}
for i in range(769, len(urls)):
    # get the link to the file you want to download from the national library
    print(i)
    url = urls[i]
    print(url)
    ScraperURL = 'https://numerique.banq.qc.ca' + url
    print(ScraperURL)
    
    driver.get(ScraperURL);
    time.sleep(10)   
    driver.switch_to.frame(driver.find_element(By.ID, "pdf_js_viewer_iframe"));
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.ID, 'download'))).click()
    time.sleep(2)
    wait_for_downloads()

    
    list_of_files = glob.glob("/Users/emily/Downloads/*.pdf")
    latest_file = max(list_of_files, key=os.path.getctime)
    
    # rename the file to be the last digits in the URL after "/"
    # does this in files, as selenium currently doesn't have the infrastructure to do this
    new_filename = url[26:]
    filename = latest_file.split("/Downloads\\")[-1].split(".")[0]
    new_path = latest_file.replace(filename, new_filename)
    os.rename(latest_file, new_path)
    
    list_of_files = glob.glob("/Users/emily/Downloads/*.pdf")
    renamed_latest_file = max(list_of_files, key=os.path.getctime)
    
    tempdict = {url : renamed_latest_file}

    print(tempdict)
    urlfiles.update(tempdict)
```
